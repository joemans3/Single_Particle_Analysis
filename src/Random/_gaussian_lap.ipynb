```python
from numpy import linspace, sqrt, sin, exp, convolve, abs, where
import numpy as np 
import matplotlib.pyplot as plt

def rect(x,radius,val):
    return where(abs(x)<=radius, val, 0)
```

```python
x = np.linspace(-100,100,1000)
#x = np.arange(-10,10,0.1)
y = rect(x,radius = 2,val = 1)
plt.plot(x,y)
```

```python
from scipy.ndimage import gaussian_filter1d
sigma = 2
print(sigma)
y_gaus = gaussian_filter1d(y, sigma=sigma,mode = 'constant',cval = 0)
plt.plot(x,y_gaus)
```

```python
from scipy import ndimage, misc

lap = (sigma**2) *ndimage.laplace(y_gaus)
plt.plot(x,lap)
print(np.max(np.abs(lap)))
```
