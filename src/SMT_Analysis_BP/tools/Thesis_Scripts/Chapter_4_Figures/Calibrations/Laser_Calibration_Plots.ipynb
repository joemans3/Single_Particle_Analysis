```python
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('svg')  
```

```python
#import the data from Laser_Box_Calibration.py
from src.SMT_Analysis_BP.tools.Thesis_Scripts.Chapter_4_Figures.Calibrations.Laser_Box_Calibration import UNITS, wavelengths,CALIBRATION

#print the UNITS
print(UNITS)
print(wavelengths)
print(CALIBRATION)
```

```python
#lets plot the calibration data for each wavelength on a single plot 
fig, ax = plt.subplots(figsize=(5,5))
for i in range(len(wavelengths)):
    ax.plot(CALIBRATION[wavelengths[i]]['percent'],CALIBRATION[wavelengths[i]]['power'],'+--',label="{0} nm".format(wavelengths[i]))
ax.set_xlabel('Percent in Software (%)')
ax.set_ylabel('Power (uW) at the sample plane')
ax.legend()
ax.set_title('Laser Power Calibration')
ax.set_yscale('log')
plt.show()
```
